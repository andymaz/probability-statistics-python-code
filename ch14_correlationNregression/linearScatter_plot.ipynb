```python
import numpy as np
import matplotlib.pyplot as plt

# Define data points
xs = np.array(range(1, 11))
ys = np.array([1.2, 3.9, 7.1, 9.5, 11.7, 13.8, 
               17.1, 20.1, 21.2, 22.8])

# Create scatter plot
plt.figure(figsize=(4, 4))
plt.scatter(xs, ys, color="navy", marker="o")
plt.title("Scatter Plot of Observed Data")
plt.xlabel("x")
plt.ylabel("y")

# Adjust layout, save, and display
plt.tight_layout()
plt.savefig("img/linearScatter_plot.png", dpi=300)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create DataFrame with x and y values
df = pd.DataFrame({'x': xs, 'y': ys})

# Add additional columns for calculations
df['xy'] = df['x'] * df['y']     # product of x and y
df['x^2'] = df['x']**2           # square of x

# Compute sums of columns
s_x = sum(df['x'])
s_y = sum(df['y'])
s_xy = sum(df['xy'])
s_xx = sum(df['x^2'])
n = len(df['x'])                 # number of data points

# Compute slope (m) using least-squares formula
def slope():
    return (n * s_xy - s_x * s_y) / (n * s_xx - s_x**2)

m = slope()
print(f'm = {m}')  # Output: m  2.457

# Compute intercept (b) using least-squares formula
def intercept():
    return (s_y - m * s_x) / n

b = intercept()
print(f'b = {b}')  # Output: b  -0.6733

# Compute fitted y-values for the regression line
y_fit = m * xs + b

# Plot scatter and fitted line
plt.scatter(xs, ys, color="red", label="Data points")
plt.plot(xs, y_fit, color="blue", label=f"Fitted line: y = {m:.3f}x + {b:.4f}")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Scatter Plot with Least-Squares Fitted Line")
plt.tight_layout()
plt.savefig('img/linear_regression.png')
plt.show()
```
