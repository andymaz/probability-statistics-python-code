```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate x-values
x = np.linspace(0, 10, 50)

# Generate quadratic data with noise
y = 0.5 * x**2 - 2*x + 5 + np.random.normal(0, 2, size=x.shape)

# Transform features for quadratic regression
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x.reshape(-1, 1))

# Fit polynomial regression
reg = LinearRegression().fit(X_poly, y.reshape(-1, 1))

# Predict y-values
y_fit = reg.predict(X_poly)

# Plot scatter and fitted curve
plt.scatter(x, y, color='green', label='Quadratic noisy')
plt.plot(x, y_fit, color='red', label='Polynomial fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Fit to Quadratic Data')
plt.legend()
plt.tight_layout()
plt.savefig('img/quadratic_regression.png')
plt.show()

# Print learned parameters
print(f"Intercept (b0) = {reg.intercept_[0]:.4f}")
print(f"Linear term (b1) = {reg.coef_[0][1]:.4f}")
print(f"Quadratic term (b2) = {reg.coef_[0][2]:.4f}")
```
