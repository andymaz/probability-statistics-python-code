```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate x-values
x = np.linspace(-5, 5, 50)

# Generate data from a cubic polynomial with noise
y = 2 * x**3 - 4*x**2 + 6*x + 2 + np.random.normal(0, 40, size=x.shape)

# Transform features for polynomial regression
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(x.reshape(-1, 1))

# Fit polynomial regression
reg = LinearRegression().fit(X_poly, y.reshape(-1, 1))

# Predict y-values
y_fit = reg.predict(X_poly)

# Scatter plot with fitted curve
plt.figure()
plt.scatter(x, y, color="blue", alpha=0.7, label="Noisy data")
plt.plot(x, y_fit, color="red", linewidth=2, label="Fitted cubic")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic Polynomial Regression")
plt.legend()
plt.show()

# Compute residuals
residuals = y - y_fit.flatten()

# Residual plot
plt.figure()
plt.scatter(x, residuals, color="blue", alpha=0.7)
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("x")
plt.ylabel("Residuals")
plt.title("Residual Plot (Cubic Regression)")
plt.show()

# Print fitted coefficients
print("Fitted coefficients (a0, a1, a2, a3):", reg.intercept_[0], *reg.coef_[0][1:])
```
