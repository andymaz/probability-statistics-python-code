```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 5
p = 1/6
x = np.arange(0, n+1)  # possible number of successes

# Binomial PMF
pmf = binom.pmf(x, n, p)

# Plot
fig, ax = plt.subplots(figsize=(6,4))
bars = ax.bar(x, pmf, color='skyblue', edgecolor='black')

# Highlight x=3
bars[3].set_color('orange')
bars[3].set_edgecolor('red')

# Labels
ax.set_xticks(x)
ax.set_xlabel("Number of successes (six)")
ax.set_ylabel("Probability")
ax.set_title("Binomial distribution B(5, 1/6) â€” Highlight x=3")
fig.savefig('joey.jpg')
plt.show()
```
