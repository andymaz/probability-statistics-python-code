```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import bernoulli

values = np.array([0, 1])
dist = bernoulli.pmf(p=0.02, k=values)  # probability mass function

fig, ax = plt.subplots(figsize=(5, 4))
sns.barplot(x=values, y=dist, ax=ax)

ax.set_xticks(values)            # set tick positions
ax.set_xticklabels(["0", "1"])   # set tick labels

ax.set_ylabel("Probability")
ax.set_title("Bernoulli Distribution (p=0.02)")

fig.savefig('img/rotten_apple.jpg')  # save figure
plt.show()                            # display figure
```
