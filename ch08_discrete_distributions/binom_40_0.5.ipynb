```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 40
p = 0.5

# Possible outcomes: integers from 0 to n
values = list(range(n + 1))

# Compute the binomial PMF for each value
dist = [binom.pmf(r, n, p) for r in values]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the PMF as a bar chart
ax.bar(values, dist, color='r', label=f'n={n} p={p}')

# Display x-axis ticks every 5 units for readability
plt.xticks(list(range(0, n + 1, 5)))

ax.set(xlabel='x', ylabel='Probability')
ax.set_title(f'Binomial PMF (n={n}, p={p})')

# Save figure before displaying
fig.savefig('img/binom_40_0.5.jpg')

plt.show()

```
