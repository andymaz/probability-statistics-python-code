```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

lambda_ = 6.8
k = 12
print(f'result = {poisson.pmf(k, lambda_):0.3f}')
```

```python

# Parameters
lambda_ = 6.8

# X values: number of named storms (0 to 19)
xs = np.arange(20)

# PMF probabilities
ys = poisson.pmf(xs, lambda_)

# Create figure
fig, ax = plt.subplots(figsize=(5, 3))

# Bar plot
sns.barplot(x=xs, y=ys, color='skyblue')

# Customise ticks and labels
ax.set_xticks(xs[::2])
ax.set_xlabel("Number of storms", fontsize=10)
ax.set_ylabel("Probability", fontsize=10)
ax.set_title("Poisson PMF: Named storms", fontsize=12)

# Save figure for LaTeX
fig.savefig('img/named_storms.png')

plt.show()
```
