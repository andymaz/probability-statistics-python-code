```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import hypergeom

# Parameters
M = 200  # total marbles
K = 20   # red marbles
n = 50   # sample size

# Values of X: number of red marbles
xs = np.arange(0, 21)

# PMF values
ys = hypergeom.pmf(xs, M, K, n)

# Create figure
fig, ax = plt.subplots(figsize=(5, 3))  # slightly wider for readability

# Bar plot
sns.barplot(x=xs, y=ys, color='skyblue', ax=ax)

# Customize ticks and labels
ax.set_xticks(xs[::2])  # every 2nd tick
ax.set_xlabel("Number of red marbles", fontsize=12)
ax.set_ylabel("Probability", fontsize=12)
ax.set_title("Hypergeometric Distribution", fontsize=14)

# Save figure for LaTeX
fig.savefig('img/hypergeom_red_marbles.jpg', bbox_inches='tight')

plt.show()
```
