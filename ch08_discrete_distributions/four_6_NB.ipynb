**Notice:** The probability of obtaining the 4th six on the 20th roll is

$$
P(\text{4th six on 20th roll}) = \text{nbinom.pmf}(16, 4, 1/6),
$$

because in the Negative Binomial distribution, the input counts the number of failures before the \(r\)-th success, so \(16\) failures + \(4\) successes = \(20\) rolls.


```python
print(nbinom.pmf(16, 4, 1/6))
```

```python
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
p = 1/6          # Probability of rolling a six
r = 4            # Number of successes
max_throws = 20  # Maximum total trials to display

# x values: number of failures
k_failures = np.arange(0, max_throws - r + 1)  # failures = total_trials - r

# PMF probabilities
ys = nbinom.pmf(k_failures, r, p)

# Total trials = failures + successes
total_trials = k_failures + r

# Plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.set_xlabel('Total number of throws')
ax.set_ylabel('Probability')
ax.set_title('Probability of obtaining the 4th six')
sns.barplot(x=total_trials, y=ys, color='skyblue')
fig.savefig('img/die_4th_six')
plt.show()
```
