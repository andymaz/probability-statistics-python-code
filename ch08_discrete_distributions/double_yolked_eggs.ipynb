```python
import numpy as np
from scipy.stats import geom
import matplotlib.pyplot as plt
import seaborn as sns

# Probability of an egg being double-yolked
p = 0.001  

# Support: first occurrence every 100 trials up to 2000
xs = np.arange(1, 2001, 100)
ys = geom.pmf(xs, p)

# Plot
fig, ax = plt.subplots(figsize=(8, 4))
sns.barplot(x=np.arange(len(xs)), y=ys, color='skyblue') 

# Show only every 2nd x-label to reduce crowding
ax.set_xticks(np.arange(0, len(xs), 2))
ax.set_xticklabels(xs[::2])

# Labels and title
ax.set_xlabel('Trial of first double-yolk')
ax.set_ylabel('Probability')
ax.set_title('Geometric distribution: double-yolked eggs')
fig.savefig('img/double_yolked_eggs.jpg')
plt.show()
```
