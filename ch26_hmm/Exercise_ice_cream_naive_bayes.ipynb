```python
# Filename: ice_cream_naive_bayes.py

# Naive Bayes approximation of Ice-Cream Parlour HMM

pi = {"Q": 0.7, "B": 0.3}
B = {
    "Q": {"L": 0.6, "M": 0.3, "H": 0.1},
    "B": {"L": 0.2, "M": 0.3, "H": 0.5}
}

observations = ["L", "H", "H", "M"]

def naive_bayes_prob(pi, B, obs):
    total = 0
    for state in pi:
        p = pi[state]
        for o in obs:
            p *= B[state][o]
        total += p
    return total

P_NB = naive_bayes_prob(pi, B, observations)
print(f"Naive Bayes probability of O: {P_NB:.6f}")
```
