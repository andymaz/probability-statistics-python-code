```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Define ReLU and its derivative
# -------------------------
def ReLU(x):
    return np.maximum(0, x)
    
def ReLU_derivative(x):
    return np.where(x > 0, 1, 0)

# -------------------------
# Generate data
# -------------------------
xs = np.linspace(-6, 6, 100)

# -------------------------
# Plot ReLU and derivative
# -------------------------
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(xs, ReLU(xs), label='ReLU', linewidth=2)
ax.plot(xs, ReLU_derivative(xs), label="Derivative", linestyle='--', linewidth=2)
ax.set_title('ReLU Activation Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.axhline(0, color='k', linewidth=1)
ax.axvline(0, color='k', linewidth=1)
ax.legend()

# -------------------------
# Save figure
# -------------------------
fig.savefig('img/reLU.png', dpi=300, bbox_inches='tight')
plt.show()
```
