```python
"""
sigmoid_and_derivative_plot.py

Generates a plot of the sigmoid activation function and its derivative.
Useful for illustrating activation functions in introductory deep learning material.
"""

import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# -------------------------
# Generate grid
# -------------------------
xs = np.linspace(-6, 6, 400)
g = sigmoid(xs)
g_prime = g * (1 - g)

# -------------------------
# Plot
# -------------------------
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(xs, g, label="Sigmoid")
ax.plot(xs, g_prime, label="Derivative")
ax.axhline(y=0, color='k', linewidth=0.8)
ax.axvline(x=0, color='k', linewidth=0.8)

ax.set_title("Sigmoid Function and Its Derivative")
ax.grid(True)
ax.legend()

# Save figure
fig.savefig("img/sigmoid.png", dpi=300)

plt.show()
```
