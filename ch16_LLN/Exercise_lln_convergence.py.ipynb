```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Parameters
# -----------------------------
n = 5000

# -----------------------------
# (a) Normal distribution
# -----------------------------
mu = 15
sigma = 36

normal_samples = np.random.normal(mu, sigma, n)
normal_running_mean = np.cumsum(normal_samples) / np.arange(1, n + 1)

plt.plot(normal_running_mean, label="Running mean")
plt.axhline(mu, linestyle="--", label="True mean")
plt.xlabel("n")
plt.ylabel(r"$\overline{X}_n$")
plt.title("LLN: Normal Distribution")
plt.legend()
plt.show()

# -----------------------------
# (b) Bernoulli distribution
# -----------------------------
p = 0.5

bernoulli_samples = np.random.binomial(1, p, n)
bernoulli_running_mean = np.cumsum(bernoulli_samples) / np.arange(1, n + 1)

plt.plot(bernoulli_running_mean, label="Running mean")
plt.axhline(p, linestyle="--", label="True mean")
plt.xlabel("n")
plt.ylabel(r"$\overline{X}_n$")
plt.title("LLN: Bernoulli Distribution")
plt.legend()
plt.show()

# -----------------------------
# (d) Cauchy distribution
# -----------------------------
cauchy_samples = np.random.standard_cauchy(n)
cauchy_running_mean = np.cumsum(cauchy_samples) / np.arange(1, n + 1)

plt.plot(cauchy_running_mean)
plt.xlabel("n")
plt.ylabel(r"$\overline{X}_n$")
plt.title("Failure of LLN: Cauchy Distribution")
plt.show()
```

### Why the Cauchy Distribution Causes Failure of the LLN

The (weak and strong) Laws of Large Numbers require the existence of a finite expectation:

$$
\mathbb{E}[X] < \infty .
$$

For a Cauchy random variable \(X\), the expectation

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f_X(x)\, dx
$$

does not exist. The integral diverges due to the heavy tails of the distribution.
