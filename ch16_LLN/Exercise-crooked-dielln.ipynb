```python
import numpy as np
import matplotlib.pyplot as plt

# Crooked die
faces = np.array([1,2,3,4,5,6])
probs = np.array([0.1,0.2,0.3,0.1,0.2,0.1])

# Simulate 5000 rolls
n = 5000
samples = np.random.choice(faces, size=n, p=probs)

# Running mean
running_mean = np.cumsum(samples) / np.arange(1, n+1)

# Plot convergence
plt.figure(figsize=(8,4))
plt.plot(running_mean, label='Running Mean')
plt.axhline(y=3.3, color='red', linestyle='--', label='Theoretical Mean')
plt.xlabel('Number of Rolls')
plt.ylabel('Running Mean')
plt.title('Convergence of Running Mean for Crooked Die')
plt.legend()
plt.show()
```
