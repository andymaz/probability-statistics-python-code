```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Logistic growth function
def logistic(t, K, A, r):
    return K / (1 + A * np.exp(-r * t))

# Generate synthetic data
np.random.seed(42)
t = np.linspace(0, 10, 50)
K_true = 100
A_true = 9
r_true = 0.5
noise = np.random.normal(0, 5, size=t.shape)
P = logistic(t, K_true, A_true, r_true) + noise

# Fit the logistic model to data
popt, pcov = curve_fit(logistic, t, P, p0=[120, 10, 0.4])
K_fit, A_fit, r_fit = popt

# Predictions from fitted model
P_fit = logistic(t, K_fit, A_fit, r_fit)

# Plot data and fitted curve
plt.figure(figsize=(8,5))
plt.scatter(t, P, color='blue', alpha=0.6, label='Noisy Data')
plt.plot(t, P_fit, color='red', lw=2, label='Fitted Logistic Curve')
plt.xlabel("Time t")
plt.ylabel("Population P(t)")
plt.title("Non-Linear Regression: Logistic Growth")
plt.legend()
plt.tight_layout()
plt.savefig('img/logistic_growth.png')
plt.show()

print(f"Fitted Parameters:\nK={K_fit:.2f}, A={A_fit:.2f}, r={r_fit:.2f}")
```
