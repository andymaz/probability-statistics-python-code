```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Logistic function
def logistic(t, K, A, r):
    return K / (1 + A * np.exp(-r * t))

# Generate synthetic data
np.random.seed(42)
t_data = np.linspace(0, 10, 50)
K_true, A_true, r_true = 100, 9, 0.8
y_true = logistic(t_data, K_true, A_true, r_true)
noise = np.random.normal(0, 3, size=t_data.shape)
y_noisy = y_true + noise

# Fit logistic model
initial_guess = [90, 5, 0.5]
params_opt, _ = curve_fit(logistic, t_data, y_noisy, p0=initial_guess)
K_est, A_est, r_est = params_opt

# Plot data and fitted curve
plt.scatter(t_data, y_noisy, label="Noisy data")
t_fine = np.linspace(0, 10, 200)
plt.plot(t_fine, logistic(t_fine, *params_opt), 'r-', label="Fitted curve")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Non-linear Regression: Logistic Growth")
plt.legend()
plt.show()

# Print estimated parameters
print(f"Estimated parameters:\nK = {K_est:.3f}, A = {A_est:.3f}, r = {r_est:.3f}")
```
