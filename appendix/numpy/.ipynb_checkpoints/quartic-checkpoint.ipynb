```python
# Plotting a polynomial function in Python using NumPy and Matplotlib.
# Demonstrates vectorised evaluation of the function, labelled axes, gridlines, and a plot title.

import numpy as np                 # Import NumPy for numerical operations
import matplotlib.pyplot as plt    # Import Matplotlib for plotting

# Define the polynomial function f(x) = x^4 + 8x^3 + 6x^2 - 11x + 5
f = lambda x: (x**4) + 8*(x**3) + 6*(x**2) - 11*x + 5

# Create an array of 100 points evenly spaced between -7 and 2
xs = np.linspace(-7, 2, 100)

# Evaluate the function at each point in the array (vectorised operation)
ys = f(xs)

# Set up the plot labels, title, and grid for clarity
plt.xlabel("x")                                         # Label x-axis
plt.ylabel("f(x)")                                      # Label y-axis
plt.title("Graph of f(x) = x^4 + 8x^3 + 6x^2 - 11x + 5")  # Plot title
plt.grid(True)                                          # Enable grid

# Plot the function
plt.plot(xs, ys)                                        # Plot y vs x
plt.tight_layout()
plt.savefig('img/quartic.png')
plt.show()                                              # Display the plot
```
