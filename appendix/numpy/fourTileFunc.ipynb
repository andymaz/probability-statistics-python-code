```python
import matplotlib.pyplot as plt
import numpy as np

# Define functions and create evaluation points
f = lambda x: x**4 + 8*x**3 + 6*x**2 - 11*x + 5
fx = np.linspace(-7, 2, 100)
fy = f(fx)

g = lambda x: 50*np.sin(x)
gx = np.linspace(-2*np.pi, 2*np.pi, 100)
gy = g(gx)

h = lambda x: np.cos(x)
hx = np.linspace(-2*np.pi, 2*np.pi, 100)
hy = h(hx)

k = lambda x: 1/x
kx = np.linspace(0.2, 6, 100)
ky = k(kx)

# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, constrained_layout=True)

# Plot each function on its respective subplot
ax[0, 0].plot(fx, fy, label='f', color='red')
ax[0, 0].set_title("Function f(x)")
ax[0, 0].legend()

ax[0, 1].plot(gx, gy, label='g', color='green')
ax[0, 1].set_title("Function g(x)")
ax[0, 1].legend()

ax[1, 0].plot(hx, hy, label='h', color='magenta')
ax[1, 0].set_title("hFunction (x)")
ax[1, 0].legend()

ax[1, 1].plot(kx, ky, label='k', color='brown')
ax[1, 1].set_title("Function k(x)")
ax[1, 1].legend()

# Save the figure and display it
fig.savefig('img/fourTileFunc.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dictionary of functions with their plotting domains
# Each entry: name  (function, (xmin, xmax, num_points))
funcs = {
    'f': (lambda x: x**4 + 8*x**3 + 6*x**2 - 11*x + 5, (-7, 2, 1000)),
    'g': (lambda x: 50*np.sin(x), (-2*np.pi, 2*np.pi, 1000)),
    'h': (lambda x: np.cos(x), (-2*np.pi, 2*np.pi, 1000)),
    'k': (lambda x: 1/x, (0.2, 6, 1000))
}

# Colours for each function plot
colours = ['red', 'green', 'magenta', 'brown']  

# Create a 2x2 grid of subplots, with automatic spacing
fig, ax = plt.subplots(2, 2, constrained_layout=True)

# Loop through functions and plot each in its subplot
for i, (name, (func, domain)) in enumerate(funcs.items()):
    r, c = divmod(i, 2)            # map index to subplot row/col
    x = np.linspace(*domain)       # generate x-values for the domain
    ax[r, c].plot(x, func(x), color=colours[i], label=name)
    ax[r, c].set_title(f"Function {name}(x)")
    ax[r, c].legend()

fig.savefig('img/fourTileFunc.png')

plt.show()
```
