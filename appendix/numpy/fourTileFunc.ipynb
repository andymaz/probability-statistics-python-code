```python
import matplotlib.pyplot as plt
import numpy as np

# Define functions and create evaluation points
f = lambda x: x**4 + 8*x**3 + 6*x**2 - 11*x + 5
fx = np.linspace(-7, 2, 100)
fy = f(fx)

g = lambda x: 50*np.sin(x)
gx = np.linspace(-2*np.pi, 2*np.pi, 100)
gy = g(gx)

h = lambda x: np.cos(x)
hx = np.linspace(-2*np.pi, 2*np.pi, 100)
hy = h(hx)

k = lambda x: 1/x
kx = np.linspace(0.2, 6, 100)
ky = k(kx)

# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2)

# Plot each function on its respective subplot
ax[0, 0].plot(fx, fy, label='f', color='red')
ax[0, 0].legend()

ax[0, 1].plot(gx, gy, label='g', color='green')
ax[0, 1].legend()

ax[1, 0].plot(hx, hy, label='h', color='magenta')
ax[1, 0].legend()

ax[1, 1].plot(kx, ky, label='k', color='brown')
ax[1, 1].legend()

# Save the figure and display it
fig.savefig('img/fourTileFunc.png')
plt.show()
```

```python
import matplotlib.pyplot as plt  # For plotting
import numpy as np               # For numerical operations

# Define four functions in a dictionary with their names as keys
funcs = {
    'f': lambda x: x**4 + 8*x**3 + 6*x**2 - 11*x + 5,
    'g': lambda x: 50*np.sin(x),
    'h': lambda x: np.cos(x),
    'k': lambda x: 1/x
}

# Define x-values for each function (ranges chosen to suit each function)
xs = [np.linspace(-7, 2, 100),
      np.linspace(-2*np.pi, 2*np.pi, 100),
      np.linspace(-2*np.pi, 2*np.pi, 100),
      np.linspace(0.2, 6, 100) ]
      
# Colours for each plot
colours = ['red', 'green', 'magenta', 'brown']  

# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2)

# Loop over the functions and plot each on the appropriate subplot
for i, (name, func) in enumerate(funcs.items()):
    r, c = divmod(i, 2)  # Compute row and column indices
    ax[r, c].plot(xs[i], func(xs[i]), color=colours[i], label=name)
    ax[r, c].legend()    # Add a legend showing the function name

# Optional: save the figure to a file
# fig.savefig('fourTileFunc.png')

# Display the plots
plt.show()
```
