```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm 

# Generate underlying normal dataset
len_normal_data = 10000
mu, sigma = 0, 1
xpoints = np.random.normal(mu, sigma, len_normal_data)

# Parameters for sampling
sample_len = 100
no_trial = 2000
means = []

# Sampling loop
for _ in range(no_trial):
    sample = [random.randint(0, len_normal_data-1) 
              for _ in range(sample_len)]
    x_bar = sum(xpoints[i] for i in sample) / sample_len
    means.append(x_bar)

_, ax = plt.subplots()

sns.kdeplot(data=xpoints, ax=ax, color='r', fill=True)
ax.set_title("Original Normal Data Distribution")
ax.set_xlabel('x')
ax.set_ylabel('Probability Density')
plt.tight_layout()
plt.savefig('img/originalNormal.png')        
plt.show()
```

```python
_, ax = plt.subplots()

sns.kdeplot(data=means, ax=ax, color='g', fill=True)
ax.set_title("Distribution of Sample Means")
ax.set_xlabel('Sample Means')
ax.set_ylabel('Probability Density')
plt.tight_layout()
plt.savefig('img/sampleMeansNormal.png')        
plt.show()
plt.show()
```
