```python
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Population
len_normal_data = 10000
mu, sigma = 0, 1
xpoints = np.random.normal(mu, sigma, len_normal_data)

# Function to create normal distribution curve
def normal_shroud(m, sgm, lng=len_normal_data):
    xs = np.linspace(m - 4*sgm, m + 4*sgm, lng)
    return xs, norm.pdf(xs, m, sgm)

# Function to compute sample means
def mean_dist(sl=1, no_trial=2000, xpoints=xpoints):
    means = []
    for _ in range(no_trial):
        sample = [random.randint(0, len_normal_data-1) for _ in range(sl)]
        x_bar = sum(xpoints[i] for i in sample)/sl
        means.append(x_bar)
    return means

# Sample sizes
sample_len = [10, 20, 40, 60, 80, 100]
means = [mean_dist(sl=x, no_trial=2000) for x in sample_len]

# Colours for plotting
sample_len_plot = [None, 20, 40, 60, 80, 100]
colours = ['green', 'magenta', 'brown', 
           'red', 'blue', 'cyan', 'orange', 'purple']

# Plotting
fig, ax = plt.subplots(3, 2, figsize=(8, 6))
for k, n in enumerate(sample_len_plot):
    if k == 0:
        # Population plot
        nx, ny = normal_shroud(0, 1)
        ax[0, 0].hist(xpoints, range=(-4,4), bins=40, 
                      color=colours[k], label='Population', 
                      density=True, histtype='bar', ec='black')
        ax[0, 0].plot(nx, ny, color=colours[(k+1)%len(colours)], lw=1.5)
        ax[0, 0].legend(loc='center left')
        continue

    # Sample means plots
    r, c = divmod(k, 2)
    nx, ny = normal_shroud(0, 1/(sample_len_plot[k]**0.5), 100)
    ax[r, c].hist(means[k], range=(-1,1), bins=40, color=colours[k], 
                  label=f'n = {sample_len_plot[k]}', 
                  density=True, histtype='bar', ec='black')
    ax[r, c].plot(nx, ny, color=colours[(k+1)%len(colours)], lw=1.5)
    ax[r, c].legend()

fig.savefig('img/sampleSizefull_withPop.png')
plt.show()
```
