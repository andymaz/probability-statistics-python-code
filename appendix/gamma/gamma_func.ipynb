```python
import numpy as np
from scipy.special import gamma, factorial
import matplotlib.pyplot as plt

# Define x values (avoid singularities at non-positive integers)
x = np.linspace(-4.0, 5.5, 2300)
y = gamma(x)

# Create figure and axis
fig, ax = plt.subplots(1, 1)

# Plot the Gamma function
ax.plot(x, y, 'r', alpha=0.6, label=r'$\Gamma(x)$')

# Plot factorial values (discrete points)
k = np.arange(1, 6)
ax.plot(k, factorial(k-1), 'o', alpha=0.6, label=r'$(x-1)!$, $x=1,2,\dots$')

# Set axis limits
ax.set_xlim(-4.0, 5.5)
ax.set_ylim(-10, 25)

# Add grid, labels, title, legend
ax.grid(True)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$\Gamma(x)$')
ax.set_title('Gamma Function and Factorial Values')
ax.legend(loc='best')

# Adjust layout and save
plt.tight_layout()
fig.savefig('img/gammasFunc.png')

plt.show()
```
