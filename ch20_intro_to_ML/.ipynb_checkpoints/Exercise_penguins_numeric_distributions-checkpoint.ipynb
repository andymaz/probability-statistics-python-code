```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Palmer Penguins dataset
penguins = sns.load_dataset("penguins")

# Drop missing values for simplicity
penguins = penguins.dropna(subset=["flipper_length_mm", "body_mass_g", "species"])

# Set seaborn style (no effect on mathematical content)
sns.set()

# ---------- Flipper length ----------
plt.figure()
sns.histplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    stat="density",
    common_norm=False,
    kde=True
)
plt.xlabel("Flipper length (mm)")
plt.ylabel("Density")
plt.title("Distribution of flipper length by species")
plt.show()

# ---------- Body mass ----------
plt.figure()
sns.histplot(
    data=penguins,
    x="body_mass_g",
    hue="species",
    stat="density",
    common_norm=False,
    kde=True
)
plt.xlabel("Body mass (g)")
plt.ylabel("Density")
plt.title("Distribution of body mass by species")
plt.show()
```
