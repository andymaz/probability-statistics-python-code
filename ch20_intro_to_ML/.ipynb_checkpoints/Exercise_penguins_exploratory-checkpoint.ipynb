```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = sns.load_dataset("penguins").dropna()

# -----------------------------------
# Feature selection: colour by species
# -----------------------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="body_mass_g",
    hue="species"
)
plt.title("Bill length vs Body mass (coloured by species)")
plt.show()

# -----------------------------------
# Alternative hue: colour by sex
# -----------------------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="body_mass_g",
    hue="sex"
)
plt.title("Bill length vs Body mass (coloured by sex)")
plt.show()
```
