```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load and clean dataset
penguins = sns.load_dataset("penguins").dropna()

# Define colour palette and features
palette = {"Male": "steelblue", "Female": "orange"}
features = ["flipper_length_mm", "body_mass_g"]

# Create side-by-side plots
fig, ax = plt.subplots(1, 2, figsize=(9, 4))

for i, y in enumerate(features):
    # Scatterplot
    sns.scatterplot(
        data=penguins,
        x="bill_depth_mm",
        y=y,
        hue="sex",
        palette=palette,
        ax=ax[i],
        alpha=0.8,
        legend=(i == 0)
    )
    
    # Fit linear regression line for each sex
    for sex_val in ["Male", "Female"]:
        subset = penguins[penguins["sex"] == sex_val]
        X = subset[["bill_depth_mm"]].values
        Y = subset[y].values
        model = LinearRegression()
        model.fit(X, Y)
        
        # Predict on a sorted range for smooth line
        X_sorted = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
        Y_pred = model.predict(X_sorted)
        
        # Plot regression line
        ax[i].plot(X_sorted, Y_pred, color=palette[sex_val], linewidth=2)

    # Set titles and labels
    ax[i].set_title(f"Bill Depth vs {y.replace('_', ' ').title()}", fontsize=10)
    ax[i].set_xlabel("Bill Depth (mm)")
    ax[i].set_ylabel(y.replace('_', ' ').title().replace('Mm', '(mm)').replace('G', '(g)'))

# Adjust layout, save figure, and display
plt.tight_layout()
plt.savefig("img/penguins-gender_regression.png", dpi=300, bbox_inches="tight")
plt.show()

```
