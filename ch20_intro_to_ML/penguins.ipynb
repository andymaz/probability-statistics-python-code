```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset and remove missing entries
penguins = sns.load_dataset("penguins").dropna()

# Define colour palette
palette = {"Adelie": "steelblue", "Chinstrap": "orange", "Gentoo": "green"}

# Pairwise visualisation coloured by species
g = sns.pairplot(penguins, hue="species", palette=palette)

# Save figure
g.savefig("img/penguins_pairplot.png", dpi=300, bbox_inches="tight")

plt.show()
```
