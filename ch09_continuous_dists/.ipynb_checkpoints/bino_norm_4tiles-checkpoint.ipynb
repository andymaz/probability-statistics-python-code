```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom

# Parameters
ps = [1/6, 2/6, 3/6, 5/6]
n = 40
xs = np.arange(n + 1)

def superimposed_normal(p, n):
    """Return mean, std dev, and normal density over support."""
    mean = n * p
    sd = (n * p * (1 - p))**0.5
    nx = np.linspace(mean - 4*sd, mean + 4*sd, 200)
    return mean, sd, nx, norm.pdf(nx, mean, sd)

# Plot settings
clr = ['red', 'green', 'magenta', 'brown']
fig, ax = plt.subplots(2, 2, figsize=(8, 4))

for i, p in enumerate(ps):
    r, c = divmod(i,2)
    m, sd, nx, ny = superimposed_normal(p, n)
    
    # Binomial PMF (bars)
    sns.barplot(ax=ax[r, c], x=xs, y=binom.pmf(xs, n, p),
                color='skyblue', edgecolor='black', label=f'p = {p:.3f}')
    
    # Normal PDF (curve)
    ax[r, c].plot(nx, ny, color=clr[i], linewidth=2)
    
    # Axis ticks
    xts = list(range(int(m - 4*sd), int(m + 4*sd) + 1, 4))
    if c == 1:
        plt.setp(ax[r, c], xticks=xts, yticks=[])
    else:
        plt.setp(ax[r, c], xticks=xts,
                 yticks=[y/100 for y in range(0, 16, 5)])
    
    ax[r, c].legend()

plt.tight_layout()
fig.savefig('img/bino_norm_4tiles.png')
plt.show()
```
