```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probability density function (PDF) for X
def pdf(x):
    # Return (x^2 + 1)/24 for 1 <= x <= 4, else 0
    return np.where((x >= 1) & (x <= 4), (x**2 + 1)/24, 0)

# Define the cumulative distribution function (CDF) for X
def cdf(x):
    # Clip x to the interval [1,4] for CDF computation
    x_clipped = np.clip(x, 1, 4)
    # Compute integral of PDF up to x (CDF)
    return (x_clipped**3/3 + x_clipped - (1**3/3 + 1)) / 24 * (x >= 1)

# Generate x values for plotting
x = np.linspace(0, 5, 500)

# Create plot
fig, ax = plt.subplots(figsize=(7,4))

# Plot PDF
ax.plot(x, pdf(x), label='PDF f(x)', color='blue', lw=2)

# Plot CDF with dashed line
ax.plot(x, cdf(x), label='CDF F(x)', color='red', lw=2, linestyle='--')

# Shade area under PDF for visualization
ax.fill_between(x, 0, pdf(x), color='blue', alpha=0.2)

# Add labels, title, legend, and grid
ax.set_xlabel('x')
ax.set_ylabel('Probability')
ax.set_title('PDF and CDF of X')
ax.legend()
ax.grid(True)

# Save figure to file
fig.savefig('img/pdf_and_cdf.png', dpi=300, bbox_inches='tight')

# Display the figure
plt.show()
```
