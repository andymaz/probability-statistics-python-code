```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
def normal_XY(m, sgm):
    """
        to create x-values ranging 4 times
        size of standard deviation on either side
        of the mean.
    """
    xs = np.linspace(m-(2.5*sgm), m+(2.5*sgm), 200)
    return xs, norm.pdf(xs, m, sgm)
```

```python
mu, sigma = 0, 1 # mean and standard deviation

nx, ny = normal_XY(mu, sigma)
```

```python
fig, ax = plt.subplots(3, 2,  figsize=(6, 4))
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.09, right=0.7, hspace=0.4,
                    wspace=0.35)

ax[0,0].plot(nx, ny, color='maroon', linewidth=0.8)
ax[0, 0].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (nx < -1),
        color= "lime",
        alpha= 0.3)
ax[0, 0].set_xticks([-2,-1, 0, 1, 2])
ax[0, 0].axhline(y=0, color='k',linewidth=0.7)
ax[0, 0].axvline(x=0, color='k',linewidth=0.7)
ax[0, 0].set_title('(a)', fontsize=10)
ax[0,0].set_xticks([])

ax[0,1].plot(nx, ny, color='magenta', linewidth=0.8)
ax[0, 1].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (nx < 1),
        color= "red",
        alpha= 0.3)
ax[0, 1].set_xticks([-2,-1, 0, 1, 2])
ax[0, 1].axhline(y=0, color='k',linewidth=0.7)
ax[0, 1].axvline(x=0, color='k',linewidth=0.7)
ax[0, 1].set_title('(b)', fontsize=10)
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])


ax[1, 0].plot(nx, ny, color='green', linewidth=0.8)
ax[1, 0].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (nx >= -1),
        color= "magenta",
        alpha= 0.3)
ax[1, 0].set_xticks([-2,-1, 0, 1, 2])
ax[1, 0].axhline(y=0, color='k',linewidth=0.7)
ax[1, 0].axvline(x=0, color='k',linewidth=0.7)
ax[1, 0].set_title('(c)', fontsize=10)
# set marking on y-axis blank
ax[1, 0].set_xticks([])
# ax[1, 0].set_yticks([])

ax[1,1].plot(nx, ny, color='red', linewidth=0.8)
ax[1,1].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (nx >= 1),
        color= "green",
        alpha= 0.3)
ax[1,1].set_xticks([-2, -1, 0, 1, 2])
ax[1,1].axhline(y=0, color='k',linewidth=0.7)
ax[1,1].axvline(x=0, color='k',linewidth=0.7)
ax[1,1].set_title('(d)', fontsize=10)
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])

ax[2,0].plot(nx, ny, color='b', linewidth=0.8)
ax[2,0].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (-1< nx) & (nx < 1),
        color= "brown",
        alpha= 0.3)
ax[2,0].set_xticks([-2, -1, 0, 1, 2])
ax[2,0].axhline(y=0, color='k',linewidth=0.7)
ax[2,0].axvline(x=0, color='k',linewidth=0.7)
ax[2,0].set_title('(e)', fontsize=10)
# ax[2,0].set_yticks([])

ax[2,1].plot(nx, ny, color='orangered', linewidth=0.8)
ax[2,1].fill_between(
        x=nx, 
        y1=ny, 
        # where= (-1.5 < nx)&(nx < 1.5),
        where= (-1.4 < nx) & (nx < -0.6) | (0.2 < nx) & (nx < 1.2),
        color= "crimson",
        alpha= 0.3)
ax[2,1].set_xticks([-2, -1, 0, 1, 2])
ax[2,1].axhline(y=0, color='k',linewidth=0.7)
ax[2,1].axvline(x=0, color='k',linewidth=0.7)
ax[2,1].set_title('(f)', fontsize=10)
ax[2,1].set_yticks([])


fig.savefig('img/six_z_probs.png')
plt.show()
```
