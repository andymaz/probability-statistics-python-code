```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm 

# Create x-axis values
x = np.arange(-5, 5, 0.01)

# List of standard deviations
sd = [0.9, 1.4, 2.2, 3.1]

# Colours for each curve
colours = ['red', 'green', 'blue', 'magenta']

# Compute normal pdf for each standard deviation
normals = [norm.pdf(x, 0, s) for s in sd]

# Create figure and axes
fig, ax = plt.subplots(figsize=(6, 4))

# Plot each normal curve
for n, c, s in zip(normals, colours, sd):
    ax.plot(x, n, label=r'$\mu=0, \sigma=' + str(s) + '$', color=c)

# Add legend, title and labels
ax.legend(title='Parameters')
ax.set_title('Normal Distributions')
ax.set_ylabel('Probability')
ax.set_xlabel('x')

# Save the figure
fig.savefig('img/normal_distributions.png')
plt.show()
```
