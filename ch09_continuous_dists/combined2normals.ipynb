```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Graphs of X, Y, and X+Y
fig, ax = plt.subplots(1, 1)

mu1, sigmaSqr1 = 140, 400
xs1 = np.linspace(80, 200, 100)
ax.plot(xs1, norm.pdf(xs1, mu1, np.sqrt(sigmaSqr1)),
        label='X: =140, =20')

mu2, sigmaSqr2 = 195, 484
xs2 = np.linspace(120, 270, 100)
ax.plot(xs2, norm.pdf(xs2, mu2, np.sqrt(sigmaSqr2)),
        label='Y: =195, =22')

muComb = mu1 + mu2
sigmaSqrComb = sigmaSqr1 + sigmaSqr2
xsc = np.linspace(250, 420, 100)
ax.plot(xsc, norm.pdf(xsc, muComb, np.sqrt(sigmaSqrComb)),
        label=f'X+Y: ={muComb}, ={np.sqrt(sigmaSqrComb):0.2f}')

plt.legend(title='Parameters', loc='best')
#fig.savefig('img/Combined2Normals.png')
plt.show()

# Probability calculations
mean, std_dev = muComb, np.sqrt(sigmaSqrComb)
print(f'P(X+Y >= 375) = {1 - norm.cdf(375, mean, std_dev):0.4f}')
print(f'P(310 <= X+Y <= 380) = {(norm.cdf(380, mean, std_dev) - norm.cdf(310, mean, std_dev)):0.4f}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Parameters: (mean, variance) for X, Y, X+Y
para = [(140, 400), (195, 484), (335, 884)]
names = ['X', 'Y', 'X+Y']
colors = ['blue', 'green', 'red']

# Create figure
fig, ax = plt.subplots(figsize=(8,5))

# Generate curves dynamically (mean  4*std)
xs = [np.linspace(mu - 4*np.sqrt(var), mu + 4*np.sqrt(var), 200) for mu, var in para]

# Plot each distribution
for i, x in enumerate(xs):
    ax.plot(x, norm.pdf(x, para[i][0], np.sqrt(para[i][1])),
            color=colors[i],
            label=f'{names[i]}: ={para[i][0]}, ={np.sqrt(para[i][1]):.1f}',
            linewidth=2)

# Labels, legend, and title
ax.set_xlabel('Value')
ax.set_ylabel('Probability Density')
ax.set_title('Normal Distributions and Their Linear Combination')
plt.legend(title='Distributions', loc='best')
plt.grid(alpha=0.3)
fig.tight_layout()
fig.savefig('img/Combined2Normals.png')
# Show figure
plt.show()


# Probability calculations
mean, std_dev = muComb, np.sqrt(sigmaSqrComb)
print(f'P(X+Y >= 375) = {1 - norm.cdf(375, mean, std_dev):0.4f}')
print(f'P(310 <= X+Y <= 380) = {(norm.cdf(380, mean, std_dev) - norm.cdf(310, mean, std_dev)):0.4f}')
```
