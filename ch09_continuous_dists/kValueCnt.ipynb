```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x-values over the support (slightly beyond to show axes clearly)
xs = np.linspace(0.35, 0.85, 500)

# Define the PDF f(x) = (25/38)*(3x + 2)
ys = (25/38)*(3*xs + 2)

# Create figure and axes
fig, ax = plt.subplots(figsize=(5, 3))

# Plot the PDF line
ax.plot(xs, ys, color='b', linewidth=1.0)  # PDF curve

# Shade the full support of the PDF
ax.fill_between(xs, ys, where=(xs>=0.4)&(xs<=0.8), color='brown', alpha=0.3)

# Shade a specific interval for probability calculation (e.g., 0.5 ≤ X ≤ 0.75)
ax.fill_between(xs, ys, where=(xs>=0.5)&(xs<=0.75), color='orange', alpha=0.4)

# Draw x=0 and y=0 axes for reference
ax.axhline(0, color='k', linewidth=0.7)
ax.axvline(0, color='k', linewidth=0.7)

# Customize ticks to highlight key points
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF of X: f(x) = (25/38)*(3x + 2), 0.4 <= x <= 0.6')


# Label axes and add a title
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF of X: f(x) = (25/38)(3x + 2), 0.4 ≤ x ≤ 0.8')

# Optional: add grid for easier visual reading
ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to prevent clipping
plt.tight_layout()

# Display the plot
plt.show()  # Renders the figure in the notebook or script

# Save high-resolution figure for the book
fig.savefig('img/k_value.png', dpi=300)  # Saves figure to file
```

```python

import numpy as np
import matplotlib.pyplot as plt

xs = np.linspace(0.35, 0.85, 500)
ys = (25/38)*(3*xs + 2)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(xs, ys, color='b', linewidth=1.0)

# Shade the full support
ax.fill_between(xs, ys, where=(xs>=0.4)&(xs<=0.8), color='brown', alpha=0.3)

# Shade the interval corresponding to probability
ax.fill_between(xs, ys, where=(xs>=0.5)&(xs<=0.75), color='orange', alpha=0.4)

ax.axhline(0, color='k', linewidth=0.7)
ax.axvline(0, color='k', linewidth=0.7)

ax.set_xticks([0, .4, .5, .75, .8, 1])
ax.set_xticklabels([0, .4, .5, .75, .8, 1]) 
ax.tick_params(axis='x', labelsize=6)  # smaller font

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF of X: f(x) = (25/38)*(3x + 2), 0.4 <= x <= 0.6')

ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
fig.savefig('img/k_value.png', dpi=300)
plt.show()


```
