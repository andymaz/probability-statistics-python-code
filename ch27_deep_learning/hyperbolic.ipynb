```python
"""
tanh_and_derivative_plot.py

Generates a plot of the hyperbolic tangent activation function (tanh)
and its derivative. Suitable for use in deep learning teaching material.
"""

import numpy as np
import matplotlib.pyplot as plt

# Hyperbolic tangent function
def tanh(x):
    return np.tanh(x)

# Derivative of tanh
def tanh_prime(x):
    return 1 - np.tanh(x)**2

# -------------------------
# Generate grid
# -------------------------
xs = np.linspace(-6, 6, 400)
y = tanh(xs)
y_prime = tanh_prime(xs)

# -------------------------
# Plot
# -------------------------
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(xs, y, label="tanh(x)")
ax.plot(xs, y_prime, label="tanh'(x)")
ax.axhline(0, color='k', linewidth=0.8)
ax.axvline(0, color='k', linewidth=0.8)

ax.set_title("Hyperbolic Tangent (tanh) and Its Derivative")
ax.grid(True)
ax.legend()

# Save figure
fig.savefig("img/hyperbolic.png", dpi=300)

plt.show()
```
