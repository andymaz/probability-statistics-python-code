```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Example: Model outputs (logits) for 3 classes
logits = torch.tensor([[2.0, 1.0, 0.1]])  # shape: (1 sample, 3 classes)

# True label: class 0 (using class index, not one-hot)
target = torch.tensor([0])

# Using nn.CrossEntropyLoss object
criterion = nn.CrossEntropyLoss()
loss1 = criterion(logits, target)
print("Loss using nn.CrossEntropyLoss():", loss1.item())

# Using functional API directly
loss2 = F.cross_entropy(logits, target)
print("Loss using F.cross_entropy():", loss2.item())
```
