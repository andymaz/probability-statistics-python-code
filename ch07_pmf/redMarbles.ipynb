This code generates a bar plot for the probability distribution of the number of red marbles drawn when selecting 3 marbles without replacement from a bag. The probabilities are pre-calculated using combinatorial methods.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Values of X (number of red marbles drawn) and their corresponding probabilities
# Scenario: drawing 3 marbles without replacement from a bag with a known composition
xs = np.array([0, 1, 2, 3])  # Possible counts of red marbles
ys = np.array([115/203, 75/203, 25/406, 1/406])  # Probabilities from combinatorial calculation

# Create bar chart of the probability distribution
fig, ax = plt.subplots(figsize=(5, 4))
sns.barplot(x=xs, y=ys,  errorbar=None, ax=ax)

# Label axes and set title
ax.set_xlabel('Number of red marbles')
ax.set_ylabel('Probability')
ax.set_title('Probability distribution: number of red marbles in 3 draws')

# Save the plot to the 'img' folder and display it
fig.savefig('img/threeRedMarbles.png')
plt.show()
```
