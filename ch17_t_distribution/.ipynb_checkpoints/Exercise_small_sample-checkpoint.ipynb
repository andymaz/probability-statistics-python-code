```python
from scipy.stats import t
import math

# Sample data
x_bar = 5.2
s = 0.8
n = 12
alpha = 0.05

# Degrees of freedom
df = n - 1

# t critical value
t_crit = t.ppf(1 - alpha/2, df=df)

# Margin of error
ME = t_crit * s / math.sqrt(n)

# Confidence interval
CI_lower = x_bar - ME
CI_upper = x_bar + ME

print(f"95% CI: ({CI_lower:.3f}, {CI_upper:.3f})")
```
