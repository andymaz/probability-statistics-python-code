```python
import numpy as np
from scipy.stats import t, norm
import matplotlib.pyplot as plt

# Degrees of freedom to compare
dfs = [1, 2, 5]
colors = ['magenta', 'blue', 'green']

# Create figure with two subplots: PDF (left), CDF (right)
fig, ax = plt.subplots(1, 2, figsize=(8, 3))

# Generate x-values
xs = np.linspace(-5, 5, 300)

# Standard normal PDF for reference
ns = norm.pdf(xs, 0, 1)
ax[0].plot(xs, ns, lw=1.7, label='Normal', color='brown')

# Plot t-distribution PDFs for different degrees of freedom
for df, c in zip(dfs, colors):
    ax[0].plot(xs, t.pdf(xs, df), lw=1.5, label=f' = {df}', color=c)

ax[0].set_xlabel('x')
ax[0].set_ylabel('Probability Density')  # more precise
ax[0].set_title('PDF')
ax[0].legend(loc='best', frameon=False)

# Plot t-distribution CDFs
for df, c in zip(dfs, colors):
    ax[1].plot(xs, t.cdf(xs, df), lw=1.5, label=f' = {df}', color=c)

ax[1].set_xlabel('x')
ax[1].set_title('CDF')   # no ylabel here
ax[1].legend(loc='best', frameon=False)

# Save and display the figure
fig.tight_layout()
fig.savefig('img/t_dist_pdf_cdf.png', dpi=300)
plt.show()
```
