```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters (, ) for the Beta distribution
parameters = [(0.7, 0.7), (6, 2), (2, 4), (3, 3), (3, 6)]

# x values in the support [0,1]
x = np.linspace(0, 1, 1000, endpoint=False)

# Create a figure with 2 subplots (tiles): one for PDF, one for CDF
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Plot PDFs ---
for a, b in parameters:
    axes[0].plot(x, beta.pdf(x, a, b), label=f'={a}, ={b}')
axes[0].set_title("Beta Distribution: PDF")
axes[0].set_xlabel("x")
axes[0].set_ylabel("Density")
axes[0].legend()

# --- Plot CDFs ---
for a, b in parameters:
    axes[1].plot(x, beta.cdf(x, a, b), label=f'={a}, ={b}')
axes[1].set_title("Beta Distribution: CDF")
axes[1].set_xlabel("x")
axes[1].set_ylabel("Cumulative probability")
axes[1].legend()

# Layout adjustment and save
plt.tight_layout()
plt.savefig("img/beta_PDF_CDF.png", dpi=300)
plt.show()
```
