```python
from scipy.stats import beta

print(f'mean = {beta.mean(8, 2)}')
# output: mean = 0.8

distrib = beta(a=8, b=2)
print(f'probability = {distrib.cdf(0.4):.4f}')
# output: probability = 0.0055
```

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

a, b = 8, 2
x = np.linspace(0, 1, 1000)  # include endpoint

y = beta.pdf(x, a, b)

fig, ax = plt.subplots(1, 1, figsize=(5, 3))
ax.plot(x, y, color='r', label=f'Beta({a},{b})', linewidth=1.0)

# Labels and title
ax.set_xlabel('Proportion of computers upgraded')
ax.set_ylabel('Probability density')
ax.set_title('Beta(8,2) Distribution for Computer Upgrades')

# Axis limits and ticks
ax.set_xlim(0, 1)
ax.set_ylim(0, 4)
ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
ax.set_yticks(np.linspace(0, 4, 5))

# Grid and legend
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()

# Save and show
plt.tight_layout()
fig.savefig('img/computer_upgrade.png', dpi=300)
plt.show()
```
