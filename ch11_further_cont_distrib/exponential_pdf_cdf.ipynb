```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt

# The scale parameter in scipy's expon is the mean, i.e. scale = 1 / lambda
lambdas = [0.5, 1, 1.5, 2]
xs = np.linspace(0, 10, 200)

# PDF
pdfs = [expon.pdf(xs, scale=1/l) for l in lambdas]
fig, ax = plt.subplots()
for l, y in zip(lambdas, pdfs):
    ax.plot(xs, y, label=r'$\lambda$ = '+f'{l:.1f}')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel(r'$f(x)$')
ax.set_title('Exponential PDF')
fig.savefig('img/exponential_pdf.png')
plt.show()

# CDF
cdfs = [expon.cdf(xs, scale=1/l) for l in lambdas]
fig, ax = plt.subplots()
for l, y in zip(lambdas, cdfs):
    ax.plot(xs, y, label=r'$\lambda$ = '+f'{l:.1f}')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel(r'$F(x)$')
ax.set_title('Exponential CDF')
fig.savefig('img/exponential_cdf.png')
plt.show()
```
