```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Shape parameter (alpha) - must be positive
alphas = [0.5, 1, 2, 3] 

# Create an array of x values
x = np.linspace(1, 5, 100)

colours = ['green', 'magenta', 'blue', 'red']

# Plot PDF and CDF
fig, ax = plt.subplots(1, 2, figsize=(7.5, 3))

for i, alpha in enumerate(alphas):
    # PDF plot
    ax[0].plot(x, stats.pareto.pdf(x, alpha), 
               label=f' = {alpha}', color=colours[i])
    
    # CDF plot
    ax[1].plot(x, stats.pareto.cdf(x, alpha), 
               label=f' = {alpha}', color=colours[i])

# Formatting the PDF plot
ax[0].set_title('Pareto Distribution PDF')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Probability Density')
ax[0].grid(True)
ax[0].legend()

# Formatting the CDF plot
ax[1].set_title('Pareto Distribution CDF')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Cumulative Probability')
ax[1].grid(True)
ax[1].legend()

plt.tight_layout(w_pad=4.0)
fig.savefig('img/pareto.png')
plt.show()
```
