```python
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt

# Define x-axis values
x = np.linspace(0, 10, 200)  # finer grid for smoother curves

# Plot the PDFs
fig, ax = plt.subplots(figsize=(8,5))  # define figure size upfront
for a in [1, 2, 4]:
# Compute Gamma PDFs for different shape parameters 'a' (alpha)
    y = stats.gamma.pdf(x, a=a, scale=1)
    ax.plot(x, y, label=f'$a={a}$, scale=1')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Gamma Distribution PDFs')
ax.legend()
ax.grid(True)

# Save the figure
fig.savefig('img/gammas.png', dpi=300)
plt.show()

```
