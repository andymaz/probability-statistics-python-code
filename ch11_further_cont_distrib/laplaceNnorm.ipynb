```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, norm
```

```python
# Parameters for the Laplace distribution
mu = 0     # Mean 
b = 10  # Different scale parameters (b)

# Generate a range of x values
data_points = 100
x = np.linspace(-60, 60, data_points)  
sd = (2 ** 0.5) * b
```

```python
def normal_shroud(m, sgm, lng=data_points):
    """
    to create x-values ranging 4 times
    size of the standard deviation on either side
    of the mean.
    """
    nx = np.linspace(m-(4*sgm), m+(4*sgm), lng)
    return nx, norm.pdf(nx, m, sgm)
```

```python
nx, ns = normal_shroud(mu, sd, lng=data_points)
# Plot the PDF for each value of b
plt.figure(figsize=(5, 3))


pdf_values = laplace.pdf(x, mu, b)
plt.plot(x, pdf_values, label=f'Laplace, b={b}')
plt.plot(nx, ns, label=f'normal', color='red')

# Add titles and labels
plt.title('Laplace and normal Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Show the grid and legend
plt.grid(True)
plt.legend()
plt.savefig('laplaceNnorm.jpg')

# Display the plot
plt.show()
```
