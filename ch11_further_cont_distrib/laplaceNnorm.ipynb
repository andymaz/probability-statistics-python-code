```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, norm

# ----------------------------
# Parameters for the Laplace distribution
# ----------------------------
mu = 0     # Location parameter (mean)
b = 10     # Scale parameter

# ----------------------------
# Generate x-values for plotting
# ----------------------------
data_points = 500
x = np.linspace(-60, 60, data_points)  # Range of x-values for Laplace PDF
sd = (2 ** 0.5) * b                    # Standard deviation for matching normal

# ----------------------------
# Function to generate normal PDF over a specified range
# ----------------------------
def normal_shroud(m, sgm, lng=data_points):
    """
    Create x-values and corresponding normal PDF values.
    The x-values range 4 standard deviations on either side of the mean.
    
    Parameters:
    m   : mean of the normal distribution
    sgm : standard deviation of the normal distribution
    lng : number of points for plotting
    
    Returns:
    nx : x-values for plotting
    ns : normal PDF values at nx
    """
    nx = np.linspace(m - 4*sgm, m + 4*sgm, lng)
    return nx, norm.pdf(nx, m, sgm)

# Generate normal distribution points
nx, ns = normal_shroud(mu, sd, lng=data_points)

# ----------------------------
# Plot Laplace and normal PDFs
# ----------------------------
plt.figure(figsize=(8, 5))

# Laplace PDF
pdf_values = laplace.pdf(x, mu, b)
plt.plot(x, pdf_values, label=f'Laplace (b={b})', color='blue', linewidth=2)

# Normal PDF
plt.plot(nx, ns, label='Normal', color='red', linestyle='--', linewidth=2)

# ----------------------------
# Add titles, labels, grid, and legend
# ----------------------------
plt.title('Laplace vs. Normal Distribution', fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)

# ----------------------------
# Save plot to file and display
# ----------------------------
plt.tight_layout()              # Adjust layout for better spacing
plt.savefig('img/laplaceNnorm.png', dpi=300)  # High-resolution save
plt.show()
```
