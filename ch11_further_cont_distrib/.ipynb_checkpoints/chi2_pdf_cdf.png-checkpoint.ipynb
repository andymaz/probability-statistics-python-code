```python
from scipy.stats import chi2 
import matplotlib.pyplot as plt
import numpy as np

ks = [2, 5, 8, 12]
dist = [chi2(k) for k in ks]
xs = np.linspace(0, 20, 400)

# --- Create side-by-side plots ---
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Left tile: PDF
for k, d in zip(ks, dist):
    axes[0].plot(xs, d.pdf(xs), label=f"k = {k}")
axes[0].set_title("Chi-squared Distributions (PDF)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("Density")
axes[0].legend(title="degrees of freedom")
axes[0].grid(alpha=0.3)

# Right tile: CDF
for k, d in zip(ks, dist):
    axes[1].plot(xs, d.cdf(xs), label=f"k = {k}")
axes[1].set_title("Chi-squared Distributions (CDF)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("Cumulative Probability")
axes[1].legend(title="degrees of freedom")
axes[1].grid(alpha=0.3)

fig.tight_layout()
fig.savefig("img/chi2_pdf_cdf.png", dpi=300)
plt.show()
```
