```python
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt

# Define x-axis values
x = np.linspace(0, 10, 200)  # fine grid for smooth curves

# Create a figure and axis object with a defined size
fig, ax = plt.subplots(figsize=(8,5))

# Plot the CDFs of Gamma distribution for different shape parameters
for a in [1, 2, 4]:
    # Compute Gamma CDF for shape parameter 'a' (alpha), scale=1
    y = stats.gamma.cdf(x, a=a, scale=1)
    ax.plot(x, y, label=f'$a={a}$, scale=1')  # LaTeX label for legend

# Label axes
ax.set_xlabel('x')         # x-axis
ax.set_ylabel('F(x)')      # CDF
ax.set_title('CDF of Gamma Distribution')  # plot title
ax.legend()                # display legend
ax.grid(True)              # add grid

# Save the figure as a PNG
fig.savefig('img/gammas_cdf.png', dpi=300)
plt.show()
```
