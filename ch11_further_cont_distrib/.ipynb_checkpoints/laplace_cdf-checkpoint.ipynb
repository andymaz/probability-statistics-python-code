```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# ----------------------------
# Parameters for the Laplace distribution
# ----------------------------
mu = 0                   # Location parameter (mean)
bs = [1, 5, 10, 15, 20]  # Scale parameters

# ----------------------------
# Generate x-values for plotting
# ----------------------------
x = np.linspace(-30, 30, 500)  

# ----------------------------
# Plot the CDF for each scale parameter
# ----------------------------
plt.figure(figsize=(7, 5))  

for b in bs:
    cdf_values = laplace.cdf(x, mu, b)
    plt.plot(x, cdf_values, label=f'b = {b}', linewidth=2)

# ----------------------------
# Add titles, labels, grid, and legend
# ----------------------------
plt.title('Laplace Distribution CDF for Different Scale Parameters', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Cumulative Probability', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Scale parameter b', fontsize=11)

# ----------------------------
# Save the figure and display
# ----------------------------
plt.tight_layout()                     # Adjust spacing
plt.savefig('img/laplace_cdf.png', dpi=300)  # High-resolution figure
plt.show()
```
