```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Parameters
mu = 0                       # Location parameter (mean)
bs = [1, 5, 10, 15, 20]      # Scale parameters (b)

# x values
x = np.linspace(-30, 30, 500)

# Plot PDFs
plt.figure(figsize=(6, 4))
for b in bs:
    pdf_values = laplace.pdf(x, mu, b)
    plt.plot(x, pdf_values, label=f'b = {b}')

# Titles, labels, legend
plt.title('Laplace Distribution: Varying Scale Parameter b')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Save and display
plt.savefig('img/laplace_pdf.png')
plt.show()
```
