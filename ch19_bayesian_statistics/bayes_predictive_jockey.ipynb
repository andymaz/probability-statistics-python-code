```python
import numpy as np
from scipy.stats import beta, binom

# Observed data
N = 12       # number of races observed
x = 9        # wins observed

# Prior (Beta)
alpha_prior = 1
beta_prior = 1

# Posterior parameters
alpha_post = alpha_prior + x
beta_post = beta_prior + (N - x)

# Monte Carlo approximation for predictive probability
n_new = 12   # number of future races
k_max = 10   # threshold

# Sample r from the posterior Beta
r_samples = beta.rvs(alpha_post, beta_post, size=100000)

# Compute P(X_new <= k_max | data) by averaging over posterior samples
P_pred = np.mean([binom.cdf(k_max, n_new, r) for r in r_samples])

print(f"Posterior predictive probability P(X_new <= {k_max}) ≈ {P_pred:.4f}")
```
