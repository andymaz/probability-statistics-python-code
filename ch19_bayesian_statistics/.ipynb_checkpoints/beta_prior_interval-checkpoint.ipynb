```python
from scipy.stats import beta
a, b = 12, 8   # example prior
p0 = 12/(a+b+0.0)
dens = beta.pdf(p0, a, b)
prob_interval = beta.cdf(0.65, a, b) - beta.cdf(0.55, a, b)
print(f'{dens:0.6f}, {prob_interval:0.6f}]')

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters
a, b = 12, 8
p = 0.6
interval = (0.55, 0.65)

# x-axis
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)

# Plot Beta PDF
fig, _ = plt.subplots (figsize =(8 ,5)) 
plt.plot(x, y, label=f'Beta({a},{b}) PDF')

# Mark height at p0
plt.vlines(p, 0, beta.pdf(p0, a, b), colors='red', linestyles='--', label=f'Height at p={p}')

# Shade interval
x_fill = np.linspace(interval[0], interval[1], 200)
plt.fill_between(x_fill, beta.pdf(x_fill, a, b), color='orange', alpha=0.4, label=f'P({interval[0]} ≤ p ≤ {interval[1]})')

# Labels
plt.xlabel('p (success probability)')
plt.ylabel('Density')
plt.title('Beta Prior with Interval and Height at p')
plt.legend()
plt.grid(True)
plt.tight_layout()
fig.savefig('img/beta_prior_interval.png')
plt.show()
```
