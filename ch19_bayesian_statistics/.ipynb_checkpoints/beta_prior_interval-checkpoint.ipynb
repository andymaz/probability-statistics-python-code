```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

a, b = 12, 8   # example prior
p0 = 12/(a+b+0.0)
dens = beta.pdf(p0, a, b)
prob_interval = beta.cdf(0.65, a, b) - beta.cdf(0.55, a, b)
print(f'{dens:0.6f}, {prob_interval:0.6f}]')


# Parameters
a, b = 12, 8
p = 0.6
interval = (0.55, 0.65)

# x-axis
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)

# Plot Beta PDF
fig, _ = plt.subplots (figsize =(8 ,5)) 
plt.plot(x, y, label=f'Beta({a},{b}) PDF')

# Mark height at p0
plt.vlines(p, 0, beta.pdf(p0, a, b), colors='red', linestyles='--', label=f'Height at p={p}')

# Shade interval
x_fill = np.linspace(interval[0], interval[1], 200)
plt.fill_between(x_fill, beta.pdf(x_fill, a, b), color='orange', alpha=0.4, label=f'P({interval[0]}  p  {interval[1]})')

# Labels
plt.xlabel('p (success probability)')
plt.ylabel('Density')
plt.title('Beta Prior with Interval and Height at p')
plt.legend()
plt.grid(True)
plt.tight_layout()
fig.savefig('img/beta_prior_interval.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Prior parameters (matching Snippet 1)
alpha, beta_param = 2, 5  

# Point of interest and interval
p0 = alpha / (alpha + beta_param)        # prior mean
interval = (0.25, 0.45)                  # example interval

# Compute density at p0 and probability of interval
dens = beta.pdf(p0, alpha, beta_param)
prob_interval = beta.cdf(interval[1], alpha, beta_param) - beta.cdf(interval[0], alpha, beta_param)
print(f'Density at p0 = {dens:0.6f}, Probability of interval {interval} = {prob_interval:0.6f}')

# Grid for plotting
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, alpha, beta_param)

# Plot Beta PDF
fig, _ = plt.subplots(figsize=(8, 5))
plt.plot(x, y, label=f'Beta({alpha},{beta_param}) PDF')

# Mark height at p0
plt.vlines(p0, 0, beta.pdf(p0, alpha, beta_param), colors='red', linestyles='--', label=f'Height at p0={p0:.2f}')

# Shade interval
x_fill = np.linspace(interval[0], interval[1], 200)
plt.fill_between(x_fill, beta.pdf(x_fill, alpha, beta_param), color='orange', alpha=0.4,
                 label=f'P({interval[0]}  p  {interval[1]})')

# Labels
plt.xlabel('p (success probability)')
plt.ylabel('Density')
plt.title('Beta Prior with Interval and Height at p0')
plt.legend()
plt.grid(True)
plt.tight_layout()
fig.savefig('img/beta_prior_interval.png')
plt.show()
```
