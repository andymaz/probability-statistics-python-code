```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Prior parameters
mu0 = 10
tau0_sq = 4
sigma_sq = 1

# Observed data
x = np.array([9.8, 10.5, 9.7, 10.2])
n = len(x)

# Posterior parameters
tau_n_sq = 1 / (n/sigma_sq + 1/tau0_sq)
mu_n = tau_n_sq * (np.sum(x)/sigma_sq + mu0/tau0_sq)
tau_n = np.sqrt(tau_n_sq)

# Plot prior and posterior densities
mu_vals = np.linspace(9, 11, 500)
prior_pdf = norm.pdf(mu_vals, loc=mu0, scale=np.sqrt(tau0_sq))
posterior_pdf = norm.pdf(mu_vals, loc=mu_n, scale=tau_n)

plt.plot(mu_vals, prior_pdf, label='Prior', linestyle='--')
plt.plot(mu_vals, posterior_pdf, label='Posterior')
plt.title(r"Prior and Posterior of $\mu$")
plt.xlabel(r"$\mu$")
plt.ylabel("Density")
plt.legend()
plt.show()
```
