```python
import numpy as np
from scipy.stats import binom, beta

# Observed data: Lara wins 8 out of 12
wins_observed = 8
N = 12

# Posterior: Beta(alpha, beta) with uniform prior
alpha_post = 1 + wins_observed
beta_post = 1 + N - wins_observed

# Monte Carlo parameters
n_samples = 100_000
r_samples = beta.rvs(alpha_post, beta_post, size=n_samples)

# Compute P(X_new <= 9) for N_new = 12
N_new = 12
threshold = 9

probabilities = binom.cdf(threshold, N_new, r_samples)
posterior_predictive = np.mean(probabilities)

print(f"Posterior predictive probability P(X_new <= 9): {posterior_predictive:.4f}")
```
