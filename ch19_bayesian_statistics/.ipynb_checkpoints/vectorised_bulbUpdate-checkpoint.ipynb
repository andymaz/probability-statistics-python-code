```python
import numpy as np

np.random.seed(42)

theta_values = np.array([0.002, 0.005, 0.007])
prior = np.array([1/3, 1/3, 1/3])

# simulate data
true_theta = 0.0046
n = 100
data = (np.random.rand(n) < true_theta).astype(int)

k = data.sum()    # sufficient statistic

likelihood = theta_values**k * (1 - theta_values)**(n - k)
unnorm = likelihood * prior
posterior = unnorm / unnorm.sum()

print("Posterior without looping:", posterior)
```
