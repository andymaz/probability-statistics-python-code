```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Posterior parameters
alpha_post = 11
beta_post = 4  # scipy uses scale = 1/beta
scale_post = 1 / beta_post

# Sample from the posterior
samples = gamma.rvs(a=alpha_post, scale=scale_post, size=100_000)

# Plot the posterior density
plt.hist(samples, bins=100, density=True, alpha=0.7, color='skyblue')
plt.title(r"Posterior density of $\lambda$")
plt.xlabel(r"$\lambda$")
plt.ylabel(r"Density")
plt.show()
```
