```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, norm

# -------------------------------
# Poisson-Gamma Example
# -------------------------------
# Fixed data
poisson_data = np.array([3, 2, 4])
n = len(poisson_data)
sum_x = np.sum(poisson_data)

# Different Gamma priors
priors_pg = [(1, 1), (2, 1), (5, 2)]  # (alpha, beta)

lambda_vals = np.linspace(0, 5, 500)

plt.figure(figsize=(10,4))
for alpha, beta in priors_pg:
    alpha_post = alpha + sum_x
    beta_post = beta + n
    post_pdf = gamma.pdf(lambda_vals, a=alpha_post, scale=1/beta_post)
    plt.plot(lambda_vals, post_pdf, label=f'={alpha}, ={beta}')
plt.title(r'Poisson-Gamma Posterior Densities')
plt.xlabel(r'$\lambda$')
plt.ylabel('Density')
plt.legend()
plt.show()

# -------------------------------
# Normal-Normal Example
# -------------------------------
# Fixed data
normal_data = np.array([9.8, 10.5, 9.7, 10.2])
n = len(normal_data)
sum_x = np.sum(normal_data)
sigma_sq = 1

# Different Normal priors
priors_nn = [(10, 4), (10, 1), (11, 0.25)]  # (mu0, tau0^2)

mu_vals = np.linspace(9, 12, 500)

plt.figure(figsize=(10,4))
for mu0, tau0_sq in priors_nn:
    tau_n_sq = 1 / (n/sigma_sq + 1/tau0_sq)
    mu_n = tau_n_sq * (sum_x/sigma_sq + mu0/tau0_sq)
    post_pdf = norm.pdf(mu_vals, loc=mu_n, scale=np.sqrt(tau_n_sq))
    plt.plot(mu_vals, post_pdf, label=f'0={mu0}, 0={tau0_sq}')
plt.title(r'Normal-Normal Posterior Densities')
plt.xlabel(r'$\mu$')
plt.ylabel('Density')
plt.legend()
plt.show()
```
