```python
import matplotlib.pyplot as plt

# Data
class_A = [55, 60, 62, 65, 68, 70, 72, 75, 80, 95]
class_B = [40, 45, 50, 55, 60, 65, 70, 75, 85, 90]

# Boxplot
plt.figure()
plt.boxplot([class_A, class_B], tick_labels=['Class A', 'Class B'])
plt.ylabel("Exam score")
plt.title("Boxplot of exam scores for two classes")
plt.show()
```
