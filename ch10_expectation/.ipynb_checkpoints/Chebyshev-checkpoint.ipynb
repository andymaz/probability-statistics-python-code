```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0       # mean
sigma = 1    # standard deviation
k = 2        # number of standard deviations

# x-axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Define bounds for shading
lower = mu - k*sigma
upper = mu + k*sigma

# Plot PDF
fig, ax = plt.subplots(figsize=(5, 3))
#plt.figure(figsize=(8,5))
plt.plot(x, y, label='PDF of X', color='blue')

# Shade region within k sigma
plt.fill_between(x, y, where=(x >= lower) & (x <= upper), 
                 color='orange', alpha=0.5, label=f'Within {k}σ')

# Mean line
plt.axvline(mu, color='red', linestyle='--', label='Mean μ')

# Mark the kσ bounds
plt.axvline(lower, color='green', linestyle='--', label=f'μ - {k}σ')
plt.axvline(upper, color='green', linestyle='--', label=f'μ + {k}σ')

# Labels
plt.title(f'Chebyshev Illustration: μ ± {k}σ')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
fig.savefig('img/Chebyshev.png')
plt.show()
```

**Figure: Chebyshev Inequality Illustration (μ = 0, σ = 1, k = 2)**

This figure shows the probability density function (PDF) of a random variable X with mean μ = 0 and standard deviation σ = 1.  

- The **orange shaded area** represents the interval [μ − kσ, μ + kσ] = [−2, 2], i.e., within 2 standard deviations of the mean.  
- The **red dashed line** marks the mean μ = 0.  
- The **green dashed lines** mark the bounds μ ± kσ = −2 and 2.  

According to **Chebyshev's inequality**, the probability that X falls within this interval is at least:

`P(|X - μ| < kσ) ≥ 1 - 1/k² = 1 - 1/4 = 0.75`

In this normal distribution, the **actual probability** within the orange region is higher than the Chebyshev bound, but the bound applies to **all distributions** with the same mean and standard deviation. This shows how Chebyshev’s inequality provides a conservative guarantee.
