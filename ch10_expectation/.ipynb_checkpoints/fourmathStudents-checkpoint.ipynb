```python
from math import comb

# Compute expected number of mathematics students in a sample of 4
result = sum([x * ((comb(7, x) * comb(5, 4-x)) / comb(12, 4)) for x in range(5)])
print(f'result = {result:.3f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import comb

# Values of X: number of mathematics students in sample
xs = np.arange(5)

# Compute corresponding probabilities using hypergeometric formula
ys = np.array([comb(7, x) * comb(5, 4-x) / comb(12, 4) for x in xs])

# Create bar plot
fig, ax = plt.subplots(figsize=(4, 3))
ax.set_xlabel('Number of Mathematics Students')
ax.set_ylabel('Probability')
sns.barplot(x=xs, y=ys, ax=ax, color='skyblue')

plt.tight_layout()  # automatically adjusts spacing to avoid overlap
# Save figure and display
fig.savefig('img/fourmathStudents.png')
plt.show()
```
