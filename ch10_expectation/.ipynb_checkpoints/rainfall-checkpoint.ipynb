```python
import numpy as np
import matplotlib.pyplot as plt

# Define the PDF with k=5 (correctly normalised)
def f_y(y, k=5):
    return (k * y**4) / (26 * 50**5) - (15 * y**2) / (26 * 50**3) + (30) / (26 * 50)

# Range for plotting (larger than the support)
y = np.linspace(-10, 60, 600)

# Evaluate PDF (nonzero only on [0,50])
pdf_vals = f_y(y)
pdf_vals_outside = np.where((y < 0) | (y > 50), 0, pdf_vals)

# Plot
fig, _ = plt.subplots(figsize=(5, 3))
plt.plot(y, pdf_vals_outside, label=r'$f_Y(y)$ with $k=5$', color='navy')

# Shade the valid part under the curve (0 to 50 mm)
y_valid = np.linspace(0, 50, 500)
plt.fill_between(y_valid, f_y(y_valid), color='skyblue', alpha=0.4)

plt.title("Rainfall Probability Density Function (Outer Hebrides Model)")
plt.xlabel("Daily rainfall (mm)")
plt.ylabel("Probability density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlim(-10, 60)  # show a bit outside the support
plt.ylim(bottom=0) # keep baseline at 0
plt.tight_layout()
fig.savefig('img/rainfall.png')
plt.show()
```
