```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x-values and y-values
xs1 = np.linspace(-31.5, -28.2, num=50)
ys1 = np.sqrt(576 / (xs1 + 33)**8)

xs2 = np.linspace(-34.5, -37.8, num=50)
ys2 = np.sqrt(576 / (xs2 + 33)**8)

# Create figure and axes
fig, ax = plt.subplots(figsize=(5, 4))

# Plot the curves
ax.plot(xs1, ys1, label='PDF (X â‰¥ -31)', color='blue')
ax.plot(xs2, ys2, label='symmetric branch (not part of PDF)', color='green')

# Add asymptote
ax.axvline(x=-33, color='red', label='asymptote x=-33')

# Labels and title
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Probability Density Function of X')

# Grid and legend
ax.grid(True)
ax.legend()

# Adjust layout and save figure
plt.tight_layout()
fig.savefig('img/green_blue.png')

# Show figure
plt.show()
```
