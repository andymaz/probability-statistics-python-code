```python
import numpy as np
import pandas as pd

# Define the distribution
data = {
    "x": [1, 2, 3, 4, 5, 6],
    "P": [0.05, 0.15, 0.25, 0.20, 0.25, 0.10]
}

df = pd.DataFrame(data)

# Compute mean
mu = (df["x"] * df["P"]).sum()

# Compute variance
variance = ((df["x"] - mu)**2 * df["P"]).sum()

# Standard deviation
std = np.sqrt(variance)

print("Mean:", mu)
print("Variance:", variance)
print("Standard deviation:", std)

# Optional: simulation
samples = np.random.choice(df["x"], size=10000, p=df["P"])
print("Sample mean:", samples.mean())
print("Sample variance:", samples.var())
print('-' * 40)
print("Sample variance:", samples.var(ddof=0))
print(df)
df.describe()
```
