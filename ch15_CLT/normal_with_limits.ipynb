```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

"""
Demonstration of a normal distribution curve with highlighted probability limits.
This script plots N(0,1) and shades the area between -1.5 and 1.5.
"""


def normal_XY(m, sgm):
    """
    Creates x-values ranging 4 times the standard deviation
    on either side of the mean.
    """
    xs = np.linspace(m - (4 * sgm), m + (4 * sgm), 100)
    return xs, norm.pdf(xs, m, sgm)

mu, sigma = 0, 1  # mean and standard deviation
nx, ny = normal_XY(mu, sigma)

# Plotting the graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(nx, ny, color='b')
ax.grid(True)
ax.set_title("Normal Distribution with Probability Limits Highlighted")

lower, upper = -1.5, 1.5

# Highlight the area of interest
ax.fill_between(
    x=nx, 
    y1=ny, 
    where=(lower < nx) & (nx < upper), 
    color='magenta', 
    alpha=0.2
)

# Add lines for axes
ax.axhline(y=0, color='k', linewidth=0.7)
ax.axvline(x=0, color='k', linewidth=0.7)

# Show and save the plot
fig.savefig('img/normal_with_limits.jpg')
plt.show()
```

```python
# Using scipy.stats.norm to calculate the probability
probability = norm.cdf(1.5) - norm.cdf(-1.5)
print(f'P(-1.5 <= Z <= 1.5) = {probability:.5f}')
```
