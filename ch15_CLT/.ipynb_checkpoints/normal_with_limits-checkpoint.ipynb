```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

"""
Demonstration of a normal distribution curve with highlighted probability limits.
This script plots N(0,1) and shades the area between -1.5 and 1.5.
"""


def normal_XY(m, sgm):
    """
    Creates x-values ranging 4 times the standard deviation
    on either side of the mean.
    """
    xs = np.linspace(m - (4 * sgm), m + (4 * sgm), 100)
    return xs, norm.pdf(xs, m, sgm)

mu, sigma = 0, 1  # mean and standard deviation
nx, ny = normal_XY(mu, sigma)

# Plotting the graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(nx, ny, color='b')
ax.grid(True)
ax.set_title("Normal Distribution with Probability Limits Highlighted")

lower, upper = -1.5, 1.5

# Highlight the area of interest
ax.fill_between(
    x=nx, 
    y1=ny, 
    where=(lower < nx) & (nx < upper), 
    color='magenta', 
    alpha=0.2
)

# Add lines for axes
ax.axhline(y=0, color='k', linewidth=0.7)
ax.axvline(x=0, color='k', linewidth=0.7)

# Show and save the plot
fig.savefig('img/normal_with_limits.png')
plt.show()
```

```python
# Using scipy.stats.norm to calculate the probability
probability = norm.cdf(1.5) - norm.cdf(-1.5)
print(f'P(-1.5 <= Z <= 1.5) = {probability:.5f}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Use LaTeX-style fonts (optional but recommended for books)
plt.rcParams.update({
    "text.usetex": False,        # set True if you compile with LaTeX
    "font.family": "serif",
    "font.size": 10
})

def normal_XY(m, sgm):
    """
    Creates x-values ranging 4 times the standard deviation
    on either side of the mean, with sufficient resolution.
    """
    xs = np.linspace(m - 4*sgm, m + 4*sgm, 1000)  # increased resolution
    return xs, norm.pdf(xs, m, sgm)

mu, sigma = 0, 1
nx, ny = normal_XY(mu, sigma)

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(nx, ny, linewidth=1.5)
ax.grid(True, linewidth=0.3)

ax.set_title("Normal Distribution with Probability Limits Highlighted")

lower, upper = -1.5, 1.5

ax.fill_between(
    nx,
    ny,
    where=(lower < nx) & (nx < upper),
    alpha=0.25,
    color='magenta'
)

ax.axhline(0, linewidth=0.7)
ax.axvline(0, linewidth=0.7)

# === IMPORTANT PART ===
# Save as VECTOR graphic
fig.savefig(
    "img/normal_with_limits.pdf",
    bbox_inches="tight"
)

plt.show()
```
