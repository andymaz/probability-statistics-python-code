```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu, sigma = 15, 36
n_trials = 5000  # number of samples

# Draw samples
samples = np.random.normal(mu, sigma, n_trials)
#print(len(samples))

# Compute running average
running_mean = np.cumsum(samples) / np.arange(1, n_trials+1)

# Plot
fig, _ = plt.figure(figsize=(7,4))
plt.plot(running_mean, label='Sample Average')
plt.axhline(mu, color='r', linestyle='--', label='True Mean μ')
plt.xlabel("Number of Samples (n)")
plt.ylabel("Running Average")
plt.title("Law of Large Numbers in Action")
plt.legend()
plt.grid(True)
fig.savefig('img/lln_visual-approching.png', dpi=300, bbox_inches='tight')
plt.show()
```
