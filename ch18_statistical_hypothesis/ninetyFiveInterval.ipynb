```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def normal_XY(m, sgm):
    """
    Create x-values covering ±4 standard deviations
    and return corresponding normal PDF values.
    """
    xs = np.linspace(m - 4*sgm, m + 4*sgm, 200)
    return xs, norm.pdf(xs, m, sgm)

mu, sigma = 0, 1  # mean and standard deviation

nx, ny = normal_XY(mu, sigma)

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(nx, ny, color='b', linewidth=0.8)
ax.fill_between(
    x=nx, 
    y1=ny, 
    where=(nx <= -1.96) | (nx >= 1.96),  # shaded tails
    color="magenta",
    alpha=0.3
)
ax.set_xticks([-4, -1.96, 0, 1.96, 4])
ax.axhline(y=0, color='k', linewidth=0.7)
ax.axvline(x=0, color='k', linewidth=0.7)
ax.set_xlabel('X')
ax.set_ylabel('Probability Density')
ax.set_title('95% significance level: shaded tails')
ax.grid(True)
plt.tight_layout()
fig.savefig('img/ninetyFiveInterval.png')
plt.show()
```
