```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def normal_XY(m, sgm):
    """
        to create x-values ranging 4 times
        size of standard deviation on either side
        of the mean.
    """
    xs = np.linspace(m-(4*sgm), m+(4*sgm), 200)
    return xs, norm.pdf(xs, m, sgm)


mu, sigma = 0, 1 # mean and standard deviation

nx, ny = normal_XY(mu, sigma)

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(nx, ny, color='b', linewidth=0.8)
ax.fill_between(
        x=nx, 
        y1=ny, 
        where= (-2 < nx)&(nx < 2),
        color= "magenta",
        alpha= 0.3)
ax.set_xticks([-4, -2, 0, 2, 4])
ax.axhline(y=0, color='k',linewidth=0.7)
ax.axvline(x=0, color='k',linewidth=0.7)
ax.set_xlabel('X')
ax.set_ylabel('Probability Density')
ax.set_title('Standard Normal Distribution')
ax.grid(True)
plt.tight_layout()
fig.savefig('img/normal_2_sigma.png')
plt.show()
```
