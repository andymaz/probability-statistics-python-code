```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu0 = 0       # H0 mean
sigma = 1
n = 30
num_samples = 10000

p_values = np.zeros(num_samples)

for i in range(num_samples):
    sample = np.random.normal(mu0, sigma, n)
    Z = (np.mean(sample) - mu0) / (sigma / np.sqrt(n))
    p_values[i] = 1 - norm.cdf(Z)  # right-tailed P-value

plt.figure(figsize=(6,3))
plt.hist(p_values, bins=50, density=True, color='skyblue', edgecolor='k', alpha=0.7)
plt.xlabel('P-value')
plt.ylabel('Density')
plt.title('Simulated P-values under $H_0$')
plt.grid(True)
plt.tight_layout()
plt.savefig('img/p_value-H0_distribution.png')
plt.show()

print(f"Mean of P-values: {np.mean(p_values):.3f}")
```
