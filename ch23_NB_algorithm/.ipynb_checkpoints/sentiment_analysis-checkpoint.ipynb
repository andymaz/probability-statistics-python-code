```python
import spacy

text = """
... Dave watched as the forest burned up on the hill,
... only a few miles from his house. The car had
... been hastily packed and Marta was inside trying to round
... up the last of the pets. "Where could she be?" he wondered
... as he continued to wait for Marta to appear with the pets.
... """

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

word_list = [token.text for token in doc if token.is_alpha]

print(word_list)
```

```python
import spacy
import glob
import os

# Load the English spaCy model once
nlp = spacy.load("en_core_web_sm")

# Define the directory path and the file extension pattern
directory_path = "./pos/"
file_extension_pattern = "*.txt"

# Use glob to find all files matching the pattern in the directory
# This returns a list of file paths (e.g., ['./pos_files_dir/review1.pos', ...])
file_paths = glob.glob(os.path.join(directory_path, file_extension_pattern))

if not file_paths:
    print(f"No files with extension '{file_extension_pattern}' found in {directory_path}")
else:
    # Iterate through each file found
    i = 1
    for file_path in file_paths:
        try:
            # Read the content of the current file
            with open(file_path, 'r', encoding='utf-8') as f:
                text = f.read()

            # Process the text with spaCy
            doc = nlp(text)

            # Filter for only alphabetic words, excluding punctuation and numbers
            words_only_list = [
                token.text for token in doc
                if token.is_alpha and not token.is_punct and not token.is_space
            ]

            print(f"\n--- Words extracted from: {file_path} ---")
            print(words_only_list)
            i += 1
            if i > 11:
                break
        except IOError as e:
            print(f"Error reading file {file_path}: {e}")
        except Exception as e:
            print(f"An unexpected error occurred during processing: {e}")

```
