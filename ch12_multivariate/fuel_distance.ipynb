```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
engine_size = np.array([1.0, 1.2, 1.5, 1.8, 2.0, 2.2, 2.5, 3.0, 3.2, 3.5])  # litres
fuel_cons = np.array([5.8, 6.1, 6.8, 7.4, 7.8, 8.1, 8.9, 9.8, 10.1, 10.7])  # L/100km

# Compute sample means
mean_X = np.mean(engine_size)
mean_Y = np.mean(fuel_cons)
print(f"Mean engine size: {mean_X:.2f} L")
print(f"Mean fuel consumption: {mean_Y:.2f} L/100km")

# Compute covariance matrix
cov_matrix = np.cov(engine_size, fuel_cons, bias=True)
print("Covariance matrix:\n", cov_matrix)

# Scatter plot
plt.figure(figsize=(4,4))
plt.scatter(engine_size, fuel_cons, color='teal', s=50)
plt.xlabel('Engine Size (L)')
plt.ylabel('Fuel Consumption (L/100km)')
plt.title('Scatter of Engine Size vs Fuel Consumption', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.savefig('img/fuel_distance.png')
plt.show()
```
