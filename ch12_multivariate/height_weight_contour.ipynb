```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# mock dataset: 1000 men, roughly realistic height/weight distribution
np.random.seed(42)
heights = np.random.normal(loc=175, scale=7, size=1000)  # cm
weights = np.random.normal(loc=90, scale=15, size=1000)  # kg

# build DataFrame
df = pd.DataFrame({
    "Height_cm": heights,
    "Weight_kg": weights
})


sns.kdeplot(
    data=df, x="Height_cm", y="Weight_kg",
    fill=True, 
    cmap="Blues",       # lighter colormap
    thresh=0.05, 
    levels= 20,          
    alpha=1.0           # slightly transparent
)

sns.kdeplot(
    data=df, x="Height_cm", y="Weight_kg",
    color="black",      # outline contours in black
    levels=15,
    linewidths=1
)

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("HeightWeight Density Contour", pad=15)
plt.tight_layout() 
plt.savefig("img/height_weight_contour.png", dpi=300)
plt.show()
```
