```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define domain (reduced grid resolution)
x = np.linspace(0, 2, 30)   # fewer points  coarser grid
y = np.linspace(2, 4, 30)
X, Y = np.meshgrid(x, y)

# Define function
Z = (2*X + Y - 6) / -4

# Plot
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor="k", alpha=0.8)

# Labels
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title("Joint PDF Example")

# Control tick density
ax.set_xticks([0, 1, 2])
ax.set_yticks([2, 3, 4])
ax.set_zticks([0, 0.5, 1])

plt.savefig("img/twoD.png", dpi=300, bbox_inches="tight")
plt.show()
```
