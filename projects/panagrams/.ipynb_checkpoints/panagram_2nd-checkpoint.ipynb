```python
import nltk
from nltk.corpus import words
from collections import Counter

# Download the word list if it's not already downloaded
#nltk.download('words')

class Pangram:
    def __init__(self, letters, mandatory):
        self.letters = letters
        self.mandatory = mandatory
        self.pool = self.letters + [self.mandatory]

    def generate_valid_words(self):
        valid_words = set()
        pool_counter = Counter(self.pool)

        # Filter dictionary: only words of length >=4 containing mandatory letter
        english_words = [w.lower() for w in words.words() if len(w) >= 4 and self.mandatory in w]

        for word in english_words:
            word_counter = Counter(word)
            # Check each letter does not exceed available count
            if all(word_counter[l] <= pool_counter.get(l, 0) for l in word):
                valid_words.add(word)

        return valid_words

def main():
    letters = [x for x in 'oarnetee']  # letter pool
    p = Pangram(letters, 'v')
    valid_words = p.generate_valid_words()

    for i, word in enumerate(sorted(valid_words), 1):
        print(f'{i:2}\t{word}')

if __name__ == '__main__':
    main()
```
