```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load Seeds dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt"
column_names = ['area', 'perimeter', 'compactness', 'length_kernel', 
                'width_kernel', 'asymmetry', 'length_groove', 'class']
seeds = pd.read_csv(url, sep=r'\s+', names=column_names)

# Standardize numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(seeds.iloc[:, :-1])
X_scaled_df = pd.DataFrame(X_scaled, columns=column_names[:-1])

# Compute correlation matrix
corr_matrix = X_scaled_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Heatmap of Seeds Dataset Features")
plt.tight_layout()
plt.savefig('img/seeds_correlation_heatmap.png')  
plt.show()
```
