```python
# -------------------------------
# SVM Regularisation Parameter C Analysis
# -------------------------------
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# Load dataset
X, y = load_iris(return_X_y=True)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Different C values
C_values = [0.01, 0.1, 1, 10, 100]
accuracies = []
support_vectors = []

# Train SVM for each C
for C_val in C_values:
    svm = SVC(kernel='linear', C=C_val, random_state=42)
    svm.fit(X_train_scaled, y_train)
    acc = svm.score(X_test_scaled, y_test)
    n_sv = len(svm.support_)
    accuracies.append(acc)
    support_vectors.append(n_sv)
    print(f"C={C_val}, Accuracy={acc:.3f}, Support Vectors={n_sv}")

# Plot Accuracy vs C
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(C_values, accuracies, marker='o')
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Test Accuracy')
plt.title('Accuracy vs C')

# Plot Support Vectors vs C
plt.subplot(1,2,2)
plt.plot(C_values, support_vectors, marker='o', color='red')
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Number of Support Vectors')
plt.title('Support Vectors vs C')

plt.tight_layout()
plt.show()
```
