```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu, sigma = 15, 36   # mean and standard deviation
n_trials = 5000      # number of random draws

# Generate random samples from N(mu, sigma^2)
samples = np.random.normal(mu, sigma, n_trials)

# Compute the running mean (sample average up to n)
running_mean = np.cumsum(samples) / np.arange(1, n_trials+1)

# Create plot
fig, ax = plt.subplots(figsize=(7,4))

# Plot running average of the samples
ax.plot(running_mean, label='Sample Average')

# Plot the true mean as a horizontal reference line
ax.axhline(mu, color='r', linestyle='--', label='True Mean Î¼')

# Labels and title
ax.set_xlabel("Number of Samples (n)")
ax.set_ylabel("Running Average")
ax.set_title("Law of Large Numbers in Action")

# Add legend and grid
ax.legend()
ax.grid(True)

# Save figure as high-resolution PNG
fig.savefig('img/lln_visual_convergence.png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
