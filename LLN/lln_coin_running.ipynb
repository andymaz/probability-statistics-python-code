```python
import numpy as np
import matplotlib.pyplot as plt

# Probability of head
p = 0.5
n_trials = 5000   # number of coin tosses

# Simulate tosses: 1 = head, 0 = tail
samples = (np.random.rand(n_trials) < p).astype(float)

# Compute running proportion of heads
running_mean = np.cumsum(samples) / np.arange(1, n_trials+1)

# Plot running proportion vs. number of tosses
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(running_mean, label='Running proportion of Heads')

# Add horizontal line for the true probability
ax.axhline(p, color='r', linestyle='--', label='True p = 0.5')

# Labels, grid, and legend
ax.set_xlabel("Number of Tosses (n)")
ax.set_ylabel("Running Proportion")
ax.set_title("LLN: Fair Coin Toss")
ax.grid(True)
ax.legend()

# Save high-resolution figure and display
fig.savefig('img/lln_coin_running.png', dpi=300, bbox_inches='tight')
plt.show()
```
