```python
# -------------------------------
# K-means Clustering on Synthetic Dataset
# -------------------------------
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data
np.random.seed(42)
cluster1 = np.random.randn(70, 2) + np.array([0, 0])
cluster2 = np.random.randn(70, 2) + np.array([5, 5])
cluster3 = np.random.randn(60, 2) + np.array([0, 5])
X = np.vstack([cluster1, cluster2, cluster3])

# Apply K-means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

# Plot clusters and centroids
plt.figure(figsize=(6,6))
for i in range(3):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i+1}')
plt.scatter(centroids[:,0], centroids[:,1], c='red', s=200, marker='X', label='Centroids')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('K-means Clustering on Synthetic Dataset')
plt.legend()
plt.show()

# Compute WCSS
wcss = kmeans.inertia_
print(f"Within-cluster sum of squares (WCSS): {wcss:.2f}")
```
