```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n = 100
trials = 1_000_000

# Generate sums and means of Uniform(0,1)
sums = [sum(np.random.rand(n)) for _ in range(trials)]
means = [np.mean(np.random.rand(n)) for _ in range(trials)]

# Plot
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

# --- (a) Distribution of sums ---
rng = (40, 60)
nx0 = np.linspace(rng[0], rng[1], 200)
ny0 = norm.pdf(nx0, loc=n/2, scale=np.sqrt(n/12))  # mean=50, sd≈2.89

ax[0].hist(sums, range=rng, bins=70, density=True, 
           color='r', edgecolor='black', alpha=0.6)
ax[0].plot(nx0, ny0, color='b', lw=2)
ax[0].set_xlabel(r'(a) Distribution of sums of $n$ i.i.d. $\mathsf{U}(0,1)$ r.v.s')
ax[0].grid()

# --- (b) Distribution of means ---
rng = (0.4, 0.6)
nx1 = np.linspace(rng[0], rng[1], 200)
ny1 = norm.pdf(nx1, loc=0.5, scale=np.sqrt(1/12/n))  # mean=0.5, sd≈0.0289

ax[1].hist(means, range=rng, bins=70, density=True, 
           color='m', edgecolor='black', alpha=0.6)
ax[1].plot(nx1, ny1, color='g', lw=2)
ax[1].set_xlabel(r'(b) Distribution of means of $n$ i.i.d. $\mathsf{U}(0,1)$ r.v.s')
ax[1].grid()

plt.tight_layout()
fig.savefig('img/sum_meanCLT.jpg', dpi=300)
plt.show()
```
