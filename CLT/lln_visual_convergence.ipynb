```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu, sigma = 15, 36
n_trials = 5000  # number of samples

# Draw samples
samples = np.random.normal(mu, sigma, n_trials)
#print(len(samples))

# Compute running average
running_mean = np.cumsum(samples) / np.arange(1, n_trials+1)

# Plot
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(running_mean, label='Sample Average')
ax.axhline(mu, color='r', linestyle='--', label='True Mean Î¼')
ax.set_xlabel("Number of Samples (n)")
ax.set_ylabel("Running Average")
ax.set_title("Law of Large Numbers in Action")
ax.legend()
ax.grid(True)
fig.savefig('img/lln_visual_convergence.png', dpi=300, bbox_inches='tight')
plt.show()
```
