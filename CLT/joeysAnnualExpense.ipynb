```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def normal_XY(mu, sigma):
    """
    Create x-values spanning Â±4 standard deviations around the mean
    and return the corresponding normal PDF values.
    """
    xs = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    return xs, norm.pdf(xs, mu, sigma)

# Parameters: number of employees and total expense stats
n = 26853
mu, sigma = 345 * n, 567 * np.sqrt(n)

# Generate normal curve for total expense
nx, ny = normal_XY(mu, sigma)

# Create figure and plot PDF
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(nx, ny, color='b', label='Total expense distribution')

# Highlight area under curve for X < 9.4 million (probability of budget sufficiency)
ax.fill_between(nx, ny, where=(nx < 9.4e6), color="magenta", alpha=0.3, label='X < 9.4M')

# Set x-axis ticks and labels
ax.set_xticks([9.0e6, 9.2e6, 9.4e6])
ax.set_xlabel("Total expense ($)")
ax.set_ylabel("Density")
ax.set_title("Distribution of total annual employee expense")
ax.legend()

# Save figure
fig.savefig('img/joeysAnnualExpense.jpg')

# Display the plot
plt.show()
```

```python
# Number of employees
n = 26853

# Total mean and standard deviation of annual expenses
mu, sigma = 345 * n, 567 * (n ** 0.5)

# Calculate probability that total expense is less than $9.4M
probability = norm.cdf(9.4e6, mu, sigma)

# Display result
print(probability)
```
